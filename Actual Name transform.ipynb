```python
def find_duplicate(x):
    for i in range(len(x)):
        if x[:i] == x[i:]:
            return x[:i]
        if x[:i-1] == x[i:]:
            return x[:i]
    else:
        return np.nan
```

```python
players["Name"] = players.player.apply(find_duplicate)
```

```python

import pandas as pd
import datetime
import unidecode
import numpy as np
import sys

year = input("Year start? ")
x = pd.read_html('http://www.foxsports.com/nba/players?teamId=0&season=%s&position=0&playerName=&country=0&grouping=0&weightclass=0' %(year))[0]
```

```python


for i in range(2,30):
    try:
        x = x.append(pd.read_html('http://www.foxsports.com/nba/players?teamId=0&season=%s&position=0&page=%s&country=0&grouping=0&weightclass=0' %(year,i))[0])
    except:
        print ("ends on page:", i)
        break

x.columns = ['Player','team','position','Height','Weight','Birthdate']

def splitter(x):
    return(x.split(",")[1])

x['Full_name'] = x.Player.apply(splitter)

x.index = [i for i in range(x.shape[0])]


def age(x):
    if x != "-":
        return ((datetime.datetime.today()-datetime.datetime.strptime(x,"%m/%d/%Y")).days)/365.25
    else:
        return np.nan


x['age'] = x.Birthdate.apply(age)

def splitspaceone(x):
    return x.split(" ")[1]
def splitspacetwo(x):
    return " ".join(x.split(" ")[3:])

x['First_name'] = x.Full_name.apply(splitspaceone)
x['Last_name'] = x.Full_name.apply(splitspacetwo)

def scrapedname(row):
        return (row['First_name'][0] +'. ' +row['Last_name'])



x = x.drop('Player',axis=1)

x['Name'] = x.apply(scrapedname,axis=1)


def conversion(x):
    if type(x) != int and type(x) != float and type(x)!=np.float64 and type(x) != np.int64:
        return unidecode.unidecode(x)
    else:
        return x
x = x.applymap(conversion)

x.to_csv("Player_list_%s.csv"%(year))
```

```python
trades = pd.read_html("https://en.wikipedia.org/wiki/List_of_2017%E2%80%9318_NBA_season_transactions")[3]
```

```python
import datetime
```

```python
def date_it(x):
    try:
        return datetime.datetime.strptime(x,'%B %d').replace(year=2017)
    except:
        return datetime.datetime(1900,1,1)
```
