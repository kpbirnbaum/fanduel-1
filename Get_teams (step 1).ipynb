```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
import re


url = 'http://espn.go.com/nba/teams'
r = requests.get(url)

soup = bs(r.text,'lxml')
tables = soup.find_all('div', class_='pl3')

```

```python
teams = []
prefix_1 = []
prefix_2 = []
teams_urls = []
for table in tables:
    extension = table.a['href']
    team = table.a.find('h2').text
    teams.append(team) 
    url = 'https://www.espn.com'+extension
    teams_urls.append(url)
    prefix_1.append(url.split('/')[-2])
    prefix_2.append(url.split('/')[-1])    
    
    
dic = {'url': teams_urls, 'prefix_2': prefix_2, 'prefix_1': prefix_1}
teams = pd.DataFrame(dic, index=teams)
teams.index.name = 'team'
teams.to_csv("teams.csv")
```
